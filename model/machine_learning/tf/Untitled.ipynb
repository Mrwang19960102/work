单变量线性回归 f(x) = ax+b    使用均方差作为成本函数，也就是说预测值和真实值之间的差的平方取均值

```python
df = pd.read_excel('./data_source/Income.xlsx')
```

```python
df.head(3)
```

```python
import numpy as np

from bokeh.plotting import figure, output_notebook, show
output_notebook()
```

```python
x = list(df['Enducation'])
y = list(df['Income'])
p = figure(title="收入与受教育年限", 
           background_fill_color="#fafafa")
p.circle(x, y, legend_label="y=ax+b",
         fill_color=None, line_color="red")
p.legend.location = "top_left"
show(p)
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1,input_shape=(1,)))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam'
             ,loss='mse')
```

```python
history = model.fit(x,y,epochs=3000)
```

```python
pre_y = model.predict(x)
```

```python
y_pre = []
for i in pre_y.tolist():
    y_pre.append(i[0])
```

```python
df['pre_income'] = y_pre
df
```
