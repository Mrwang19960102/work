```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df = pd.read_excel('./data_source/test.xlsx')
df
```

```python
x = df[['a','b']]
y = df['y']
```

```python
model=tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1,input_shape=(2,)))
model.summary()
```

```python
model.compile(optimizer='adam'
             ,loss='mse')
```

```python
history = model.fit(x,y,epochs=5000)
```

```python
model.predict(x)
```

```python
y_pre = []
for i in model.predict(x).tolist():
    y_pre.append(i[0])
```

```python
y_pre
```

```python
df['pre_y'] = y_pre
df
```
